# Prepare for the prediction set

```python
import pandas as pd
from openpyxl import load_workbook
import numpy as np
from sklearn.model_selection import train_test_split
pd.set_option('display.max_rows', None)

import time 
```

# Import the data 

```python
data = pd.read_excel('new_concept_2022-10.xlsx', engine = 'openpyxl')
data = data[['platform','spu_id','spu_name']]
data.head()
```

```python
print("The length of the smaple: " + str(len(data)))
#dfall.to_excel('关键词all.xlsx')
dis_spu =data['spu_id'].astype(str).unique()
print("The number of the distinct spu_id: " + str(len(dis_spu)))
```

```python
# 去掉商品名称里的空字段；
X_predict = data['spu_name']
for i in range(len(X_predict)):
    X_t = X_predict[i].replace(' ', '')
    X_predict[i] = X_t
```

```python
# 创建存储label的空集
def init_list_of_objects(size):
    list_of_objects = list()
    for i in range(0,size):
        list_of_objects.append( list() ) #different object reference each time
    return list_of_objects

labels = init_list_of_objects(len(X_predict))

# 按照每一个spu_name的长度创建一个存储的空集label；
i=0
for X_t in X_predict:
    #print(X_t)
    labels[i]= ['']*len(str(X_t)) #test=X_train[0]
    i=i+1
   
```

```python
# label "o"
for i in range(len(labels)):
    labels[i] = ['O' if j=='' else j for j in labels[i]]
# label "sep"
for i in range(len(labels)):
    labels[i].insert(len(labels[i]),'')
```

```python
#示例
print(labels[90])
#print(X_train[1])
```

```python
start = time.time()
```

```python

start = time.time()
formatted = pd.DataFrame(columns={'char', 'cat'})
i=0
for X_t in X_predict:
    #print(X_t)
    sentence = X_t
    separated_sen = list(str(sentence).encode("utf-8").decode("utf-8"))
    separated_sen.insert(len(separated_sen),'')
    sen_labels = list(labels[i])
    for char, label in zip(separated_sen, sen_labels):
        data = [{'char': char,'cat': label}]
        # print(data)
        formatted = formatted.append(data,sort=False)
    i=i+1
end = time.time()
duration = end - start
print('Time spending in this process is: ' + str(duration))
```

```python
predict['char'] = predict['char'].astype(str)
predict['cat'] = predict['cat'].astype(str)
```

```python
for index, row in predict.iterrows():
    if row['char'] == 'nan':
        predict.at[index, 'cat'] = ''
        predict.at[index, 'char'] = ''
```
